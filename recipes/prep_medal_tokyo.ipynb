```python
import pandas as pd
```

## First pass

```python
medals = pd.read_csv('../extras/medals_tokyo_raw.csv')
medals.head()
```

```python
medals_recode = {
    'Country': 'country',
    'Gold Medal': 'gold',
    'Silver Medal': 'silver',
    'Bronze Medal': 'bronze'
}
medals = medals[list(medals_recode.keys())]
medals.rename(columns=medals_recode, inplace=True)
medals
```

## Appending 3-letter ISO code

```python
ioc_countries = pd.read_csv('../extras/ioc_countries.csv').set_index('country')
ioc_countries = ioc_countries.to_dict()
country_recode = ioc_countries.get('code')
```

```python
medals['code'] = medals.country.replace(country_recode)

assert medals.code.isna().sum() == 0
```

## Output prep

```python
medals = medals[['code'] + list(medals_recode.values())]
medals.head()
```

```python
medals.to_csv("../datasets/medal_countries_tokyo.wide.csv")
```

```python
medals_long = (
    medals
    .melt(
        id_vars=['code', 'country'],
        var_name = 'medal',
        value_name = 'quantity'
    )
)
```

```python
medals.to_csv("../datasets/medal_countries_tokyo.long.csv")
```
